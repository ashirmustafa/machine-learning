```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
df_linear = pd.read_csv("linear_data.csv")
df_nonlinear = pd.read_csv("nonlinear_data.csv")
df_mixed = pd.read_csv("mixed_data.csv")

# Plot the data
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

def plot_data(ax, df, title):
    ax.scatter(df["Feature"], df["Target"], color='b', alpha=0.6)
    ax.set_title(title)
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")

plot_data(axes[0], df_linear, "Linear Relationship")
plot_data(axes[1], df_nonlinear, "Non-Linear Relationship")
plot_data(axes[2], df_mixed, "Mixed Relationship")

plt.tight_layout()
plt.show()
```
